Red points show how `im2` and `im1` are related. Magenta points explain how `im1` and `im3` are related.

__NOTE to self:__ Before you get caught out, notice that the point pairs are in X, Y format, not in R, C format. In this convention, X points run along columns from left to right and Y points run along rows from top to bottom.

```python
# Load data
data = sio.loadmat('PracticalData.mat', squeeze_me = True);
# Unpack data

# im1 is center image with grey background. im1 has enough space to insert im2 and im3.
im1 = data['im1']
im1copy = data['im1']
# im2 is left image 
im2 = data['im2']

# pts1 and pts2 are matching points between image1 and image2
pts1 = data['pts1']
pts2 = data['pts2']


# im3 is right image
im3 = data['im3']
# pts1b and pts3 are matching points between image 1 and image 3
pts1b = data['pts1b']
pts3 = data['pts3']

# Show images and points
plt.imshow(im1)
plt.plot(pts1[0,:],pts1[1,:],'r.')
plt.plot(pts1b[0,:],pts1b[1,:],'m.')
plt.axis('off')
plt.show()

plt.imshow(im2)
plt.plot(pts2[0,:],pts2[1,:],'r.')
plt.axis('off')
plt.show()

plt.imshow(im3)
plt.plot(pts3[0,:],pts3[1,:],'m.')
plt.axis('off')
plt.show()
```

```python
def solveAXEqualsZero(A):
    [U,L,Vh]=np.linalg.svd(A,full_matrices=True)
    V=np.transpose(Vh)
    [x, y]=np.shape(V)
    h=np.zeros([x,1])
    h=V[:,y-1]
    return h
```

```python
def calcBestHomography(pts1Cart, pts2Cart):
    
   # This function should apply the direct linear transform (DLT) algorithm to calculate the best 
    # homography that maps the cartesian points in pts1Cart to their corresonding matching cartesian poitns 
    # in pts2Cart.
    
    # This function calls solveAXEqualsZero. Make sure you are wary of how to reshape h into a 3 by 3 matrix. 
    
    n_points = pts1Cart.shape[1]
    
    
    H = np.identity(3)
    
    # First convert points into homogeneous representation
    # Hint: we've done this before  in the skeleton code we provide.
    pts1Hom = np.concatenate((pts1Cart, np.ones((1,pts1Cart.shape[1]))), axis=0)
    pts2Hom = np.concatenate((pts2Cart, np.ones((1,pts2Cart.shape[1]))), axis=0)
    # Then construct the matrix A, size (n_points * 2, 9)
    # Consult the notes!
    A = np.zeros([n_points*2,9])
    
    for i in range(n_points):
               
        A[i*2,:]=[0,0,0,-pts2Hom[0,i],-pts2Hom[1,i],-1,pts1Hom[1,i]*pts2Hom[0,i],pts1Hom[1,i]*pts2Hom[1,i],pts1Hom[1,i]]
        A[((i*2)+1),:]=[pts2Hom[0,i],pts2Hom[1,i],1,0,0,0,-pts1Hom[0,i]*pts2Hom[0,i],-pts1Hom[0,i]*pts2Hom[1,i],-pts1Hom[0,i]]
    # Solve Ah = 0 using solveAXEqualsZero and get h.
    
    h = solveAXEqualsZero(A)
    # Reshape h into the matrix H, values of h go first into rows of H
    print(h.shape)
    H = np.reshape(h,(3,3))
    #H[0,:]=[h[0,],h[1,],h[2,]]
    #H[1,:]=[h[3,],h[4,],h[5,]]
    #H[2,:]=[h[6,],h[7,],h[8,]]
    return H
   
```

```python
# MAIN PROCESS TADA!!: Calculate homography from pts1 to pts2 

startTime = time.perf_counter()
h = calcBestHomography(pts2,pts1)

[pixelr1, pixelc1, col1]= np.shape(im1)
[pixelr2, pixelc2, col2]= np.shape(im2)

for y in range(pixelr1):
    for x in range(pixelc1):
        estHom = np.matmul(h,[x,y,1])
        
        [u,v,z]= estHom;
        
        u=int(np.round(u/z))
        v=int(np.round(v/z))
        
        if u<pixelc2:
            if v<pixelr2:
                im1[y,x,:]=im2[v,u,:];

print("Stitching image 2 to image 1 took: {0:.2f}s".format(time.perf_counter() - startTime))
```

```python
# Plot the intermediate panorama
plt.imshow(im1)
plt.axis('off')
plt.show()
```

Here in the above figure the stitching is tiled. I tried figuring out the constraints for preventing that but couldn't. I found the homography of matching points from img1 to img2 in that order and replaced the pixels that mapped to img2 with the corresponding matching positions in img2

```python
# TO DO: Repeat the above process mapping image 3 to image 1.

# Calculate homography from pts1b to pts3

startTime = time.perf_counter()
h2 = calcBestHomography(pts3,pts1b)

[pixelr1, pixelc1, col1]= np.shape(im1)
[pixelr3, pixelc3, col3]= np.shape(im3)

for j in range(pixelr1):
    for i in range(pixelc1):
        estHom2 = np.matmul(h2,[i,j,1])
        
        [u1,v1,z1] = estHom2
        
        u1 = int(np.round(u1/z1))
        v1 = int(np.round(v1/z1))
#        if u1<0:
#            break
#        if v1<0:
#            break
        if u1<pixelc3 and u1>=0:
            if v1<pixelr3 and v1>=0:
                im1[j,i,:]=im3[v1,u1,:]
        
print("Stitching image 3 to image 1 took: {0:.2f}s".format(time.perf_counter() - startTime))
```

```python
# Plot the final panorama
plt.imshow(im1)
plt.axis('off')
plt.show()
```

Here I found the homography for moving from img1 pixels to img3 using the matching points, went over the pixels in img1 to see if they mapped to any pixels in img3 within its boundaries and copied over the value.
